```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("Data Set Aquisition.csv")
df
```

### Data Cleaning

```python
df.info()
```

#### Checking if there are any duplicate entries

```python
duplicate_rows = df[df.duplicated()]
print('Number of duplicate Samples : ',len(duplicate_rows))
```

#### Removing the duplicates entries

```python
df = df.drop_duplicates()
df.reset_index(inplace=True)
df = df.drop(['index'],axis =1)
```

```python
df
```

#### Removing non important Columns

```python
df = df.drop(["Timestamp","Your Good Name?"],axis=1)
```

### Label Encoding

```python
list(df["Organization?"].value_counts().index)
```

```python
def f1(x):
    l = list(df["Organization?"].value_counts().index)
    return l.index(x)
df["Organization?"] = df["Organization?"].apply(f1)
```

```python
df["Last Degree?"].value_counts().index
```

```python
def f2(x):
    l = ['BSC (CS or SE)', 'MS(CS or SE)', 'Phd (CS or SE)']
    return l.index(x)
df["Last Degree?"] = df["Last Degree?"].apply(f2)
```

```python
df["Job Experience ?"].value_counts().index
```

```python
def f3(x):
    l = ['Education', 'Software Industry', 'Other',
       'Software Industry;Education', 'Education;Other',
       'Software Industry;Education;Other', 'Software Industry;Other']
    return l.index(x)
df["Job Experience ?"] = df["Job Experience ?"].apply(f3)
```

### Train Test Split

```python
df
```

```python
y = df["Job Experience ?"]
df = df.drop(["Job Experience ?","Last Degree?","Organization?"],axis=1)
```

### Standardization

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
data = scaler.fit_transform(df)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    data, y, test_size=0.20, random_state=42)
```

### Machine Learning Algorithms

```python
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
```

### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression

lg =  LogisticRegression(random_state=0)
lg.fit(X_train,y_train)

y_pred = lg.predict(X_train)
print("Training Results")

print("Training Accuracy : ",accuracy_score(y_train,y_pred))
print("Training Precision : ", precision_score(y_train,y_pred,average='micro'))
print("Training Recall : ",recall_score(y_train,y_pred,average='micro'))
print("Training F1-Score : ", f1_score(y_train,y_pred,average='micro'))
print(" ")
y_pred = lg.predict(X_test)

print("Testing Results")
print("Testing Accuracy : ",accuracy_score(y_test,y_pred))
print("Testing Precision : ", precision_score(y_test,y_pred,average='micro'))
print("Testing Recall : ",recall_score(y_test,y_pred,average='micro'))
print("Testing F1-Score : ", f1_score(y_test,y_pred,average='micro'))
```

### Decision Tree Classifier


```python
from sklearn.tree import DecisionTreeClassifier

dt =  DecisionTreeClassifier(random_state=0)
dt.fit(X_train,y_train)

y_pred = dt.predict(X_train)
print("Training Results")

print("Training Accuracy : ",accuracy_score(y_train,y_pred))
print("Training Precision : ", precision_score(y_train,y_pred,average='micro'))
print("Training Recall : ",recall_score(y_train,y_pred,average='micro'))
print("Training F1-Score : ", f1_score(y_train,y_pred,average='micro'))
print(" ")
y_pred = dt.predict(X_test)

print("Testing Results")
print("Testing Accuracy : ",accuracy_score(y_test,y_pred))
print("Testing Precision : ", precision_score(y_test,y_pred,average='micro'))
print("Testing Recall : ",recall_score(y_test,y_pred,average='micro'))
print("Testing F1-Score : ", f1_score(y_test,y_pred,average='micro'))
```

### KNN Classifier

```python
from sklearn.neighbors import KNeighborsClassifier
neigh = KNeighborsClassifier(n_neighbors=5)
neigh.fit(X_train, y_train)

y_pred = neigh.predict(X_train)
print("Training Results")

print("Training Accuracy : ",accuracy_score(y_train,y_pred))
print("Training Precision : ", precision_score(y_train,y_pred,average='micro'))
print("Training Recall : ",recall_score(y_train,y_pred,average='micro'))
print("Training F1-Score : ", f1_score(y_train,y_pred,average='micro'))
print(" ")
y_pred = neigh.predict(X_test)

print("Testing Results")
print("Testing Accuracy : ",accuracy_score(y_test,y_pred))
print("Testing Precision : ", precision_score(y_test,y_pred,average='micro'))
print("Testing Recall : ",recall_score(y_test,y_pred,average='micro'))
print("Testing F1-Score : ", f1_score(y_test,y_pred,average='micro'))
```
